## 1 . Missing completly at random
Missing completely at random (MCAR) refers to a situation where the probability of data being missing is independent of both observed and unobserved data. In other words, the missingness of the data is completely random and does not depend on any specific characteristics or values in the dataset. This means that the missing data does not introduce any bias into the analysis, and the remaining data can be considered a random sample of the original dataset. MCAR is one of the most ideal scenarios for handling missing data, as it allows for unbiased estimates and valid statistical inferences.

## data is not related to any observed or unobserved data

## 2 . Missing at random
Missing at random (MAR) refers to a situation where the probability of data being missing is related to the observed data but not the unobserved data. In other words, the missingness of the data can be explained by the observed variables in the dataset. For example, if we have a dataset with information about individuals' ages and incomes, and we find that income data is more likely to be missing for older individuals, then the missingness of income is related to age (an observed variable). In this case, we can use the observed data (age) to model and account for the missing data (income), which allows for more accurate analysis and inference compared to MCAR.

## data is related to observed data but not unobserved data
## hindi me batao
Missing at random (MAR) का मतलब है कि डेटा के गायब होने की संभावना उन देखे गए डेटा से संबंधित होती है, लेकिन उन अनदेखे डेटा से नहीं होती है। दूसरे शब्दों में, डेटा के गायब होने की वजह को डेटासेट में देखे गए वेरिएबल्स से समझा जा सकता है। उदाहरण के लिए, अगर हमारे पास एक डेटासेट है जिसमें व्यक्तियों की उम्र और आय की जानकारी है, और हम पाते हैं कि आय का डेटा अधिकतर पुराने व्यक्तियों के लिए गायब होता है, तो आय का गायब होना उम्र (एक देखे गए वेरिएबल) से संबंधित होता है। इस स्थिति में, हम देखे गए डेटा (उम्र) का उपयोग करके गायब डेटा (आय) को मॉडल कर सकते हैं और उसे ध्यान में रखकर विश्लेषण और निष्कर्ष निकाल सकते हैं, जो MCAR की तुलना में अधिक सटीक होगा। 


## 3 . Missing not at random
Missing not at random (MNAR) refers to a situation where the probability of data being missing is related to the unobserved data itself. In other words, the missingness of the data is not random and cannot be explained by the observed variables in the dataset. For example, if we have a dataset with information about individuals' health status and we find that individuals with poor health are more likely to have missing health data, then the missingness of health data is related to the unobserved variable (health status). In this case, the missing data can introduce bias into the analysis, and it may require more complex methods to handle and account for the missingness in order to make valid inferences.

## data is related to unobserved data itself
## hindi me batao
Missing not at random (MNAR) का मतलब है कि डेटा के गायब होने की संभावना उन अनदेखे डेटा से संबंधित होती है। दूसरे शब्दों में, डेटा के गायब होने की वजह यादृच्छिक नहीं होती है और डेटासेट में देखे गए वेरिएबल्स से समझाई नहीं जा सकती है। उदाहरण के लिए, अगर हमारे पास एक डेटासेट है जिसमें व्यक्तियों की स्वास्थ्य स्थिति की जानकारी है, और हम पाते हैं कि खराब स्वास्थ्य वाले व्यक्तियों के स्वास्थ्य डेटा अधिकतर गायब होता है, तो स्वास्थ्य डेटा का गायब होना अनदेखे वेरिएबल (स्वास्थ्य स्थिति) से संबंधित होता है। इस स्थिति में, गायब डेटा विश्लेषण में पक्षपात (bias) ला सकता है, और इसे संभालने और गायब होने की वजह को ध्यान में रखकर सही निष्कर्ष निकालने के लिए अधिक जटिल तरीकों की आवश्यकता हो सकती है।

## Imputation techniques for handling missing values include:
1. Mean Value Imputaion
2. Median Value Imputation
3. Mode Value Imputation





```python
## 1 Mean Value Imputaion


df.age.isnull().sum()
```

```python
## distribution of titanic data age column
sns.histplot(df.age, kde=True)
```

```python
sns.displot(df.age, kde=True)
```

```python
df['Age_mean'] = df.age.fillna(df.age.mean())
```

```python
df[['age', 'Age_mean']]
```

```python
## this technique is not good for skewed data because mean is affected by outliers and skewed data has outliers
## this technique is good for normally distributed data because mean is not affected by outliers and normally distributed data does not have outliers
```

```python
## meadian value imputation
## if we have skewed data then we will use median value imputation because median is not affected by outliers and skewed data has outliers

df['Age_meadian'] = df['age'].fillna(df['age'].median())
```

```python
df[['Age_median','Age_mean', 'age']]
```
